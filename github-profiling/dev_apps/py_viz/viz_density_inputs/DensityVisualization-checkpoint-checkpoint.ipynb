### <center>Variant Density Visualization</center>

Input data *1000g_interesting_positions* is a supplied csv file. *1000g_interesting_positions* is processed from _1000Genomics_ VCF data. The csv file has 4 columns: <b>position</b>, <b>num_intersect</b>, <b>num_ref_block_cell</b> and <b>num_spos_aligned</b>

1. position - bp position
2. num_intersect - number of samples intesect at this position
3. num_ref_block_cell - numbet of intesected positions that have reference
4. num_spos_aligned - number of positions that are aligned at the position

The number of variant is <code>num_variant = num_ref_block_cell - num_intersect</code>. In the other words, the intersected positions not defined in references.

The input file 1000g_interesting_positions contains 48604666 positions. About 3% (1486310) of them has variants. 

We divide data into dense and sparse judged by the mean an number of variants. In our data, the mean value of number variant is **42.45** 

* <em><strong>Dense Area</strong></em> contains <code>num_variant >= mean</code>
* <em><strong>Sparse Area</strong></em> contains <code>num_variant &lt; mean</code> 

After the seperatio, we have distribution of number of variant per positions as illustrated in the following figure 
<center>![Image](output_7_0.png)</center>
<center><span style="color:green">**Dense**</span> vs <span style="color:red">**Sparse**</span></center>

The dense area contains 12.22% of positions and 95.80% of total variant range, the sparse area contains 87.78% of positions and 4.20% of total variant range.

```python
plt.close('all')
fig = plt.figure(figsize=(20, 15))

plt.scatter(dense.position, dense.num_variant, s=0.5, color='green', label='Dense')
plt.scatter(sparse.position, sparse.num_variant, s=0.5, color='red', label='Sparse') 
plt.legend()
plt.show()
```

During the profiling, it appears to me that many bp positions has 1 variant in 1000 samples. The following figures show the distribution of number bp position per number of variant.

```python
def max_min(mydict):
    min_v = (None, 800000)
    max_v = (None, 0)
    for k, v in mydict.items():
        if v > max_v[1]:
            max_v = (k, v)
        if v < min_v[1]:
            min_v = (k, v)
    return max_v, min_v
gn = df.groupby(df.num_variant)
nv_all = {x : len(gn.get_group(x)) for x in gn.groups }
print("All: max and min number are in (#variant, #pos) format:", max_min(nv_all))
```

The plot for dense area. The mean values is mean=42.448.

```python
nv_dense = {x : len(gn.get_group(x)) for x in gn.groups if x >= mean_val}
percent_num_pos = sum(nv_dense.values())/float(len(df.index)) * 100
percent_num_variant = len(nv_dense)/float(len(gn.groups)) * 100
print("Dense area contains %.2f%% of positions that has variants, %.2f%% of total variant range." % \
      (percent_num_pos, percent_num_variant))
max_val, min_val = max_min(nv_dense)
print("The max# of position is %d at #variant %d, and min# is %d at %d" % (max_val[1],max_val[0],min_val[1],min_val[0]) )
```

```python
plt.close('all')
fig = plt.figure(figsize=(20, 15))
plt.plot(list(nv_dense.keys()), list(nv_dense.values()), color='navy', label='Number of Positions per Dense Variants')
plt.fill_between(list(nv_dense.keys()), list(nv_dense.values()), color='lightgray')
plt.legend()
plt.show()
```

The plot for sparse area. The mean values is mean=42.448.

```python
# defined above gn = df.groupby(df.num_variant)
nv_sparse = {x : len(gn.get_group(x)) for x in gn.groups if x < mean_val}
percent_num_pos = sum(nv_sparse.values())/float(len(df.index)) * 100
percent_num_variant = len(nv_sparse)/float(len(gn.groups)) * 100
print("Sparse area contains %.2f%% of positions that has variants, %.2f%% of total variant range." % \
      (percent_num_pos, percent_num_variant))
max_val, min_val = max_min(nv_sparse)
print("The max# of position is %d at #variant %d, and min# is %d at %d" % \
      (max_val[1],max_val[0],min_val[1],min_val[0]) )
```

```python
plt.close('all')
fig = plt.figure(figsize=(20, 15))
# ax.patch.set_facecolor('navy')
plt.plot(list(nv_sparse.keys()), list(nv_sparse.values()), 'r')
plt.bar(list(nv_sparse.keys()), nv_sparse.values(), color='g', \
        label='Number of Positions per Number of Variants that Below the Mean')
plt.legend()
plt.show()
```
