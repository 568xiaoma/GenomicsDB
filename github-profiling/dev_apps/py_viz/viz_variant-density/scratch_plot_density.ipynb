```python
def get_density_pos(df_all, region_pos, low=0, high=float("inf")):
    ''' df_all - dataframe of intersect file from KG
        num_pos - list of number pos per region
        low - lowest pos (inclusive), default 0, 
        high - high pos (exclusive), default float("inf") (to the end)
        return list of num_region each as [([pos], [#var]),...,([pos], [#var])]
        example: pick 2, 3 position for dense, middle and sparse
        [ [([13416, 13379], [22, 8]), ([13416, 13379, 13503], [22, 8, 5])], [..mid reg.], [..sparse..] ]
    ''' 
    in_range = df_all[(df_all.position >= low) & (df_all.position < high)]
    mean_val = in_range.num_variant.mean()
    def mid_pos(df, nh, nl):
        return pd.concat([df[df['num_variant']>=mean_val].tail(nh), df[df['num_variant']<mean_val].head(nl)])
    sorted_df = in_range.sort_values(by=['num_variant', 'position'], ascending=[False, True])
    print("mean_val=%.2f, pos_range(%s, %s), #var_range=(%s, %s)  #df=%s" % (mean_val, low, high, \
            sorted_df['num_variant'].min(), sorted_df['num_variant'].max(), sorted_df.shape))
    all_region = list()
    all_region.append([sorted_df.head(n)[['position', 'num_variant']] for n in region_pos[0]])
    all_region.append([mid_pos(sorted_df, hn, ln)[['position', 'num_variant']] for hn, ln in [(int(n/2+n%2), int(n/2)) for n in region_pos[1]]])
    all_region.append([sorted_df.tail(n)[['position', 'num_variant']] for n in region_pos[2]])
    return all_region
```

```python
def merge_pos_list(merged_list, new_pos_list):
    ''' from [([pos], [n_var]),..] => [([[arr1_pos, [arr2_pos]], [max_nvar, min_nvar]),..]'''
    is_first = len(merged_list) == 0
    for i, (pos, n_var) in enumerate(new_pos_list):   #dl is dense[1,10..], sparse[]
        if is_first:
            merged_list.append(([pos], [max(n_var), min(n_var)]))
        else:
            merged_list[i][0].append(pos)
            ln_var = merged_list[i][1] + n_var
            merged_list[i][1][0] = max(ln_var)
            merged_list[i][1][1] = min(ln_var)
            
def merge_to_list(merged_list_item, pos, n_var):
    ''' from [([pos], [n_var]),..] => [([[arr1_pos, [arr2_pos]], [max_nvar, min_nvar]),..]'''
    merged_list_item[0].append(pos)
    merged_list_item[1][0] = max(merged_list_item[1][0], max(n_var))
    merged_list_item[1][1] = min(merged_list_item[1][1], min(n_var))
```

```python
class MyPlotter(PlotGenerator):
    @staticmethod
    def prep4plot_data(data_set):
        flat_data = [[d, p, s, [wc_in_sec(x) for x in v]] for (d, p, s), v in data_set.items()]
        df = pd.DataFrame(flat_data, columns=['density', 'pos', 'seg_size', 'result'])
        df['wc_mean'] = df.result.apply(np.mean)
        df['wc_std'] = df.result.apply(np.std)
        return df
  
    @staticmethod
    def get_exp_name(exp_name):
        i = exp_name.find('_')
        return (exp_name[0:i].upper(), exp_name[i+1:])
    
    def generate_stacked_graph(self, input_list, my_exp_list, show_graph=False):
        ''' inputs = [{'intermediate 100 position(s)' : ['1:49.23', '1:51.34', '1:48.13', '1:50.39', '1:48.48']},..]
            exp_list ['e-5_var_only', 'e-8_zlib',...]
        '''
        assert len(input_list) == len(my_exp_list) 
        assert len(input_list) == 2
        exp_list = [self.get_exp_name(x) for x in my_exp_list]
        
        for i, (btm_result_fn, btm_data) in enumerate(input_list[0].items()):
            _, top_data = list(input_list[1].items())[i]
            df_1 = self.prep4plot_data(btm_data)
            df_2 = self.prep4plot_data(top_data)

            if self.positions is None:
                self.positions = df_1.pos.unique()                 # [1, 10, 100, 1000]
                self.seg_sizes = df_1.seg_size.unique()            # [500, 1000, 3000, 6000, 9000]
                self.alphas = super().calc_alpha(len(self.seg_sizes))       # [1.0, 0.4, 0.8, 0.2, 0.6] ?
                self.ind = np.arange(len(self.positions))
            self.plot_names.clear()
            self.rects.clear()
            plt.close('all')
            _, ax = plt.subplots(1, 1, figsize=(self.fig_w, self.fig_h))
            self.calc_bars(df_2, self.top_colors[i % 2], exp_list[1][0])
            self.calc_bars(df_1, self.colors[i % 2], exp_list[0][0])

            node = os.path.basename(btm_result_fn).split('_')[0]
            title = "%s @ %s" % (' vs '.join(["%s[%s]" % st for st in exp_list]), node)
            self.make_plot(ax, title)
            if show_graph:
                plt.show()
```

```python
colors = ["lightgreen", "orange", "cyan", "red", "yellow"]

def plot_data(node_name, density_arr, index):
    ind = np.arange(len(index))
    width = 0.2
    plt.close('all')
    fig, ax = plt.subplots(1, 1, figsize=(10, 8))
    ax.set_title('Title', fontsize=16)
    rects = []
    plot_names = []
    for i, (dname, ddate) in enumerate(density_arr.items()):
        avg, std = zip(*ddate)
        ind_pos = ind + (i * width)
        print(avg, std, ind_pos)
        rects.append(plt.bar(ind_pos, avg, width, align='center', color=colors[i], yerr=std))
        plot_names.append(dname[0].upper() + dname[1:])
        
    # add some text for labels, title and axes ticks
    ax.set_xlabel('Number Positions')
    ax.set_ylabel('Wall Clock in sec')
    ax.set_title('Profiling Result at %s' % node_name, fontsize=16)

    ax.set_xticks(ind + width / 2)
    ax.set_xticklabels(index)
    trects_0 = tuple([r[0] for r in rects])
    ax.legend(trects_0, tuple(plot_names))
    
    def autolabel(myrects):
        """
        Attach a text label above each bar displaying its height
        """
        for rect in myrects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., 1.00*height,
                    '%d' % int(height),
                    ha='center', va='bottom')

    [autolabel(r) for r in rects]
    #ax.legend((rects1[0], rects2[0]), ('Dense', 'Sparse'))
    # plt.savefig('%s.png' % node_name)
    plt.show()
```
